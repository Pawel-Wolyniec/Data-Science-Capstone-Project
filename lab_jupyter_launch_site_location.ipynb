```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster, MousePosition
from math import sin, cos, sqrt, atan2, radians

# Function to calculate distance
def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance

# Sample data for demonstration
data = {
    'Launch Site': ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Lat': [28.562302, 28.563197, 28.573255, 34.632834],
    'Long': [-80.577356, -80.576820, -80.646895, -120.610745],
    'class': [1, 0, 1, 0]
}

# Create DataFrame
spacex_df = pd.DataFrame(data)

# Create a new column 'marker_color' based on the 'class' value
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Display the DataFrame to verify the 'marker_color' column
print(spacex_df)

# Create a folium map centered at a specific location
site_map = folium.Map(location=[28.562302, -80.577356], zoom_start=5)

# Create a MarkerCluster
marker_cluster = MarkerCluster().add_to(site_map)

# Add markers to the map with colors based on the 'class' value
for index, record in spacex_df.iterrows():
    # Create a Marker object with its coordinate
    marker = folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color='white', icon_color=record['marker_color']),
        popup=record['Launch Site']
    )
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Add marker_cluster to the current site_map
site_map.add_child(marker_cluster)

# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=5,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)

# Define points of interest (e.g., a railway, a city, etc.)
points_of_interest = [
    {'name': 'Railway Station', 'Lat': 28.5721, 'Long': -80.6480},
    {'name': 'City Center', 'Lat': 28.5383, 'Long': -81.3792}
]

# Calculate and display the distances
for point in points_of_interest:
    print(f"Distances from {point['name']}:")
    for index, site in spacex_df.iterrows():
        distance = calculate_distance(point['Lat'], point['Long'], site['Lat'], site['Long'])
        print(f" - {site['Launch Site']}: {distance:.2f} km")

# Display the map
site_map
```
